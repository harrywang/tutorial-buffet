```python
# dribbble image and color palette
# https://dribbble.com/shots/12236793-Breakfast
from PIL import Image
Image.open('test.png')
```

```python
import colorgram

# specify number of colors to extract from an image.
num_of_colors = 6

# extraction
colors = colorgram.extract('test.png', num_of_colors)

# examples from the repo
# colorgram.extract returns Color objects, which let you access
# RGB, HSL, and what proportion of the image was that color.
first_color = colors[0]
rgb = first_color.rgb # e.g. (255, 151, 210)
hsl = first_color.hsl # e.g. (230, 255, 203)
proportion  = first_color.proportion # e.g. 0.34

# RGB and HSL are named tuples, so values can be accessed as properties.
# These all work just as well:
red = rgb[0]
red = rgb.r
saturation = hsl[1]
saturation = hsl.s
```

```python
# proportion of the colors sorted
colors
```

```python
# build the palette

import numpy as np
color_list = []
for c in colors:
    color_list.append(list(c.rgb))

palette = np.array(color_list)  # convert to np array for plotting
palette
```
